Reading an excel file using pandas in python and speciying index, header & encoding 

```python
# Pandas as a library tends to create an index column by default while reading a file.
# By specifying index_col =0 we are specifying the pandas to use 0th column (i.e Sr no.) as the index column
# By specifying header =0 we are specifying the pandas to use the default header if we use header = None, it will demote the header
# By specifying encoding="utf-8" we are specifying the pandas to eliminate any error in case of special characters.

data = pd.read_excel("mtcars.xlsx",index_col=0,header=0,encoding="utf-8")
```

```python
data
```
